We start the project by importing the Pandas and Numpy libraries

```python
import pandas as pd
import numpy as np
```

Now lets download and read the data:

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
# I am using the exact same method to download the data as I don't know whether the code would work if I use the downloaded data file concrete_data.csv
concrete_data.head()  #To return the first 5 rows of the data for verification
```

Now we are going to divide the data into predictors and targets. We have 8 parameters; Cement, Blast Furnace Slag, Fly ash, water, superplasticizer, coarse aggregrate, fine aggregate and age as predictors and the concrete compressive strength as target.

```python
concrete_features = concrete_data.columns
predictors = concrete_data[concrete_features[concrete_features != 'Strength']] #All columns except strength
target = concrete_data ['Strength']
```

```python
predictors.head() #Check
```

```python
target.head() #Check
```

Now I am going to create a variable to have the no of predictors in the dataset. I will need this while creating the model.

```python
num_predictors = predictors.shape[1]
```

PART A of the Question
Build the model with one hidden layer of 10 node and a Relu activation function

First we import Keras and the necessary packages

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

Now we build the neural network here

```python
def regression_conc_model():
    model = Sequential()
    model.add(Dense(10, activation = 'relu', input_shape = (num_predictors,))) #10 nodes on hidden layer, relu activation function
    model.add(Dense(1)) #output layer
    
    #We are compiling the model; set adam as optimizer and mean squared error as loss function
    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    return model
    
```

Now we are going to split the data into training and test set. First we create the model

```python
model = regression_conc_model()
```

Part A1: we are going to split data using helper function from scikit-learn

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train, y_test =  train_test_split(predictors, target, test_size = 0.3, random_state = 42)
```

Now we will train and test the model with fit.

```python
#fit the model
model.fit(X_train, y_train , epochs=50, verbose=2)
```

Now we are going to evaluate the model in the test data

```python
validation_loss = model.evaluate(X_test, y_test) #Evaluating
validation_loss
```

we are going to calculate the mean squared error between predicted and actual concrete strength. We start by importing the mean_squared_error function from scikit-Learn

```python
from sklearn.metrics import mean_squared_error
```

```python
predicted_strength = model.predict(X_test) #Calculating the predicted strength which we need to calculate the MSE
```

Calculating the mean square error between predicted and actual concrete strength

```python
mse_predicted_actual = mean_squared_error(y_test, predicted_strength)
```

```python
print(mse_predicted_actual)
```

To calculate mean and standard deviation, 

```python
mean = np.mean(mse_predicted_actual) #to calculate mean
standard_deviation = np.std(mse_predicted_actual) #to calculate standard deviation
```

We are going to train the network 50 times and calculate the mean squared error and create a list of the 50 MSE's.

```python
total = 50  #no of iterations
mse_of_all_iterations = [] #empty list to store all 50 mse

for i in range(0, total):
    X_train, X_test,y_train, y_test =  train_test_split(predictors, target, test_size = 0.3, random_state = 42) #test/train split
    model.fit(X_train, y_train , epochs=50, verbose=2) #fit the model
    validation_loss = model.evaluate(X_test, y_test) #Evaluating
    predicted_strength = model.predict(X_test) #Calculating the predicted strength which we need to calculate the MSE
    mse_predicted_actual = mean_squared_error(y_test, predicted_strength) #to calculate the MSE
    mse_of_all_iterations.append(mse_predicted_actual) #to append to the list


mse_of_all_iterations = np.array(mse_of_all_iterations) #updating the list
mean = np.mean(mse_of_all_iterations) #to calculate mean
standard_deviation = np.std(mse_of_all_iterations) #to calculate standard deviation

    
```

printing out the final mean and standard devation,

```python
print('Mean:', mean )
print('Standard of Deviation:', standard_deviation )
```
